```python
#Import dependencies
import pandas as pd
import requests
import numpy as np
import pprint
import sqlalchemy
from config import token

# Python SQL toolkit and Object Relational Mapper
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
from sqlalchemy import create_engine, inspect, func
```

```python
#Import list of cities in the US that will be used for the API calls
cities_csv = pd.read_csv("1000-cities.csv")
cities_df = pd.DataFrame(cities_csv, columns = ["Rank", "City", "State", "Growth From 2000 to 2013", "Population", "Latitude", "Longitude"])
dropped = cities_df.drop(["Rank", "Growth From 2000 to 2013", "Population"], axis = 1)
lat = dropped["Latitude"]
long = dropped["Longitude"]
cities = dropped["City"]
city1 = "New York"
dropped
```

```python
#DO NOT RUN
#INFO WAS EXPORTED TO A CSV AND CALLED AGAIN

#Creating new dataframe to store information from API calls
air_quality_df = pd.DataFrame(columns = ["City", "AQI", "IDX", "IAQI", "Dominant Pollutant", "Date"])
i= 1


#Looping through cities to get Air Quality data
for city in cities:
    print(f"Processing City {i} | {city}")
    try:
            response = requests.get(f"https://api.waqi.info/feed/{city}/?token={token}").json()
            city_name = response['data']['city']['name']
            aqi = response['data']['aqi']
            idx = response['data']['idx']
            iaqi = response['data']['iaqi']['w']['v']
            pollutant = response['data']['dominentpol']
            date = response['data']['time']['iso']
            air_quality_df = air_quality_df.append({"City": city,
                                                    "AQI": aqi,
                                                    "IDX": idx,
                                                    "IAQI": iaqi,
                                                    "Dominant Pollutant": pollutant,
                                                    "Date": date}, ignore_index = True)
    except:
        print("City not found. Skipping...")
        pass
    i += 1    
            
```

```python
#First look at API dataframe
air_quality_df
```

```python
#Export dataframe to CSV
air_quality_df.to_csv('air_quality.csv', index=False)
```

```python
#Reading dataframe back in 
path1 = ("air_quality.csv")
clean_air_quality_df=pd.read_csv(path1)

#Importing Population csv
cities_csv = pd.read_csv("1000-cities.csv")
clean_cities_df = pd.DataFrame(cities_csv, columns = ["Rank", "City", "State", "Growth From 2000 to 2013", "Population", "Latitude", "Longitude"])
```

```python
#Merging both CSV files
merged_df = pd.merge(clean_air_quality_df, clean_cities_df, on = "City")
merged_df
```

```python
#Drop duplicates
clean_merged_df = merged_df.drop_duplicates()
```

```python
#Data cleanup: single out columns desired and creating new dataframe
new_clean_merged_df = clean_merged_df.loc[:,["City", "State", "Population", "AQI", "IDX", "Dominant Pollutant", "Growth From 2000 to 2013"]]
new_clean_merged_df
```

```python
#Renaming columns
aclean_merged_df = new_clean_merged_df.rename(columns = {"Growth From 2000 to 2013":"Population Growth"})
aclean_merged_df
```

```python
#Creating engine and connection
engine = create_engine(f"postgres://postgres:{password}@localhost:5432/airquality_db")
conn = engine.connect()
```

```python
#Make sure connection was working
engine.table_names()
```

```python
#Appending tables to PostgreSQL
aclean_merged_df.to_sql( name = 'AQI_Metrics', con=engine, if_exists ='append')
```

```python
#Appending second table
AQI_df = aclean_merged_df.loc[:,["City","Population", "AQI",  "Population Growth"]]
AQI_df.to_sql( name = 'AQI_City', con=engine, index = False)
```

```python
#Appending third table
IDX_df = aclean_merged_df.loc[:,["City","Population", "IDX",  "Population Growth"]]
AQI_df.to_sql( name = 'IDX_City', con=engine, index = False)
```
